```python
import random
import numpy as np

transaction_type = ['l','d']
transactions = []
n = 0
def inputFunction(input):
  global transactions
  global n
  with open('input.txt', 'r') as file:
      lines = file.readlines()
      n = int(lines[0])
      for i in range(1, n+1):
          t, s = lines[i].split()
          transactions.append((t, int(s)))
      return transactions

inputFunction('input.txt')

reg = []
for i in transactions:
  reg.append(i)


print('Register:',reg)

def fitnessFunc(register):  ### it Works!!!!

  sum = 0
  for i in register:
    if i[0] == 'd':
      sum += i[1]
      # print('deposit')
      # print(transaction[1])
    elif i[0] == 'l':
      # print('lent')
      # print(transaction[1])

      sum -= i[1]
    # if sum == 0:
    #     return len(register) - register.count(('z', 0))
    # else:
    #     return 0
  return sum


inputFunction('input.txt')
cashList = []
print('Fitness calculation:',fitnessFunc(reg))

#print('end')

chromosome = []
for i in reg:
  if i[0] == 'd':
    chromosome.append(i[1])
  else:
    chromosome.append(i[1]*-1)


print('Chromosome:',chromosome)
child1 = []
#child2 = []

def crossover(parent):
  global child1
  crossoverPoint = random.randint(1, len(parent)-1)
  firstHalf = parent[:crossoverPoint]
  secondHalf = parent[crossoverPoint:]

# concatenate the two parts in reverse order
  child1 = secondHalf + firstHalf
  return child1


  
print('Crossover:',crossover(chromosome))

def mutate(x):
    max_val = max(x)
    #print('Max',max_val)
    random_index = random.randrange(0, len(x))
    mutation = random.randint(0,max_val)
    for i in range(0, max_val+1):
      count = 0
      if count == 0:
        x[random_index] = mutation
        count += 1
      else:
        break
    return x

muT = mutate(child1)
print('mutated chromosome after crossover:',muT)

column = n
row = column
minimum = min(chromosome)
maximum = max(chromosome)

#### Genetic Algorithm  
randomPopulace = np.random.randint(minimum, maximum, (row, column))

print('Randomly Generated Registers:')
print(randomPopulace)

```
