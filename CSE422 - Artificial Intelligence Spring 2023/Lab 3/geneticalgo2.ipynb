```python
import random

transaction_type = ['l', 'd']

def inputFunction():
    with open('input.txt', 'r') as file:
        lines = file.readlines()
        n = int(lines[0])
        transactions = []
        for i in range(1, n + 1):
            t, s = lines[i].split()
            transactions.append((t, int(s)))
        return transactions

def fitnessFunc(register):
    sum = 0
    for transaction in register:
        if transaction[0] == 'd':
            sum += transaction[1]
        elif transaction[0] == 'l':
            sum -= transaction[1]
    if sum == 0:
        return len(register) - register.count(('z', 0))
    else:
        return 0

def crossover(p1, p2):
    if len(p1) != len(p2):
        return f'Both parents must be equal in length!'
    else:
        length = len(p1)
        crossoverPoint = random.randint(1, length - 1)
        child1 = p1[:crossoverPoint] + p2[crossoverPoint:]
        child2 = p2[:crossoverPoint] + p1[crossoverPoint:]
        return child1, child2

def mutate(register):
    index = random.randint(0, len(register) - 1)
    mutationType = transaction_type[random.randint(0, 1)]
    mutationValue = random.randint(1, 100000)
    register[index] = (mutationType, mutationValue)
    return register

def generateRandomRegister(numTransactions):
    register = []
    for i in range(numTransactions):
        transactionType = transaction_type[random.randint(0, 1)]
        transactionValue = random.randint(1, 100000)
        register.append((transactionType, transactionValue))
    return register

def runGeneticAlgorithm(populationSize, numGenerations):
    transactions = inputFunction()
    population = [generateRandomRegister(len(transactions)) for i in range(populationSize)]

    for i in range(numGenerations):
        fitnessScores = [fitnessFunc(register) for register in population]
        populationFitness = list(zip(population, fitnessScores))
        populationFitness.sort(key=lambda x: x[1], reverse=True)
        if populationFitness[0][1] != 0:
            return ''.join(['1' if t in populationFitness[0][0] else '0' for t in transactions])
        else:
            newPopulation = [populationFitness[0][0]]
            for j in range(populationSize - 1):
                parents = random.choices(populationFitness[:int(populationSize/2)], k=2)
                child1, child2 = crossover(parents[0][0], parents[1][0])
                child1 = mutate(child1)
                child2 = mutate(child2)
                newPopulation.extend([child1, child2])
            population = newPopulation

    return -1



output = runGeneticAlgorithm(100, 1000)
print(output)
```
