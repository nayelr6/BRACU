```python
import random

def parse_input(filename):
    with open(filename, 'r') as file:
        lines = file.readlines()
        n = int(lines[0])
        transactions = []
        for i in range(1, n+1):
            t, s = lines[i].split()
            transactions.append((t, int(s)))
        return transactions

def fitness(register):
    # Calculate the sum of non-zero elements in the register
    return sum([s for t, s in register if s != 0])

def crossover(parent1, parent2):
    # Perform single-point crossover
    n = len(parent1)
    crossover_point = random.randint(1, n-1)
    child1 = parent1[:crossover_point] + parent2[crossover_point:]
    child2 = parent2[:crossover_point] + parent1[crossover_point:]
    return child1, child2

def mutate(register, mutation_rate):
    # Randomly change some elements in the register
    for i in range(len(register)):
        if random.random() < mutation_rate:
            t, s = register[i]
            if t == 'l':
                s = -s
            register[i] = ('d', s)
    return register

def generate_random_register(n):
    # Generate a random register with n transactions
    register = []
    for i in range(n):
        t = random.choice(['l', 'd'])
        s = random.randint(1, 100000)
        if t == 'l':
            s = -s
        register.append((t, s))
    return register

def find_zero_balance_register(transactions, population_size=100, elite_size=10, mutation_rate=0.01, num_generations=1000):
    # Initialize the population with random registers
    population = [generate_random_register(len(transactions)) for i in range(population_size)]

    # Iterate over generations
    for generation in range(num_generations):
        # Evaluate the fitness of each register in the population
        fitnesses = [fitness(register) for register in population]

        # Find the elite registers with highest fitness
        elite_indices = sorted(range(population_size), key=lambda i: fitnesses[i], reverse=True)[:elite_size]
        elite_registers = [population[i] for i in elite_indices]

        # Select parents for crossover using roulette wheel selection
        fitness_sum = sum(fitnesses)
        parent_probabilities = [fitness / fitness_sum for fitness in fitnesses]
        parents = [random.choices(population, weights=parent_probabilities, k=2) for i in range(population_size - elite_size)]

        # Perform crossover and mutation to generate offspring
        offspring = []
        for i in range(population_size - elite_size):
            child1, child2 = crossover(parents[i][0], parents[i][1])
            offspring.append(mutate(child1, mutation_rate))
            offspring.append(mutate(child2, mutation_rate))

        # Replace the old population with the new population (elite + offspring)
        population = elite_registers + offspring

        # Check if any register balances out to zero
        for register in population:
            if fitness(register) == 0:
                return ''.join(['1' if s != 0 else '0' for t, s in register])

    # If no register balances out to zero, return -1
    return '-1'

# Example usage:
transactions = parse_input('input.txt')
result = find_zero_balance_register(transactions)
print(result)
```
