```python
import random

# read the input from file
with open('input.txt', 'r') as f:
    n = int(f.readline().strip())
    transactions = [f.readline().strip().split() for i in range(n)]

# convert the amounts to integers
transactions = [(t[0], int(t[1])) for t in transactions]
print(transactions)
# define the genome length
genome_length = len(transactions)

# define the fitness function
def fitness(genome):
    balance = 0
    for i in range(genome_length):
        if genome[i] == 1:
            print(genome[i])
            if transactions[i][0] == 'l':
                balance += transactions[i][1]
            else:
                balance -= transactions[i][1]
    return abs(balance)
print(fitness(transactions))
# define the crossover function
def crossover(parent1, parent2):
    crossover_point = random.randint(0, genome_length-1)
    offspring = parent1[:crossover_point] + parent2[crossover_point:]
    return offspring

# define the mutation function
def mutate(genome, mutation_rate):
    mutated_genome = []
    for bit in genome:
        if random.random() < mutation_rate:
            mutated_genome.append(1 - bit)
        else:
            mutated_genome.append(bit)
    return mutated_genome

# define the population size and mutation rate
population_size = 100
mutation_rate = 0.01

# create the initial population
population = [[random.randint(0, 1) for i in range(genome_length)] for j in range(population_size)]

# run the genetic algorithm for a fixed number of iterations
max_iterations = 1000
for iteration in range(max_iterations):
    # evaluate the fitness of each genome in the population
    fitness_scores = [fitness(genome) for genome in population]
    
    # check if a perfect solution is found
    if 0 in fitness_scores:
        print('Perfect solution found in iteration', iteration+1)
        best_genome = population[fitness_scores.index(0)]
        for i in range(genome_length):
            if best_genome[i] == 1:
                print(i+1, transactions[i][0], transactions[i][1])
        break
    
    # select the parents for reproduction
    parent_indices = random.sample(range(population_size), 2)
    parent1 = population[parent_indices[0]]
    parent2 = population[parent_indices[1]]
    
    # create the offspring genome
    offspring_genome = []
    
    # perform crossover
    offspring_genome = crossover(parent1, parent2)
        
    # perform mutation
    offspring_genome = mutate(offspring_genome, mutation_rate)
        
    # add the offspring genome to the population
    population.append(offspring_genome)
        
    # replace the least fit genome with the offspring genome
    fitness_scores.append(fitness(offspring_genome))
    least_fit_index = fitness_scores.index(max(fitness_scores))
    population.pop(least_fit_index)
    fitness_scores.pop(least_fit_index)
    
    # print the best fitness score in this iteration
    print('Iteration', iteration+1, '- Best fitness:', min(fitness_scores))
    
# if no perfect solution is found, print -1
if 0 not in fitness_scores:
    print('No perfect solution found')
```
