```python
### Lab 2 Alpha beta pruning


import random

# Function to calculate the achieved point by applying alpha-beta pruning
def alpha_beta_pruning(node, depth, alpha, beta, maximizingPlayer, total_points_to_win):
    if depth == 3: # Terminal Node
        return node
    if maximizingPlayer: # Maximizing player (Optimus Prime)
        bestValue = -float('inf')
        for child in node:
            value = alpha_beta_pruning(child, depth+1, alpha, beta, False, total_points_to_win)
            bestValue = max(bestValue, value)
            alpha = max(alpha, bestValue)
            if beta <= alpha:
                break
        return bestValue
    else: # Minimizing player (Megatron)
        bestValue = float('inf')
        for child in node:
            value = alpha_beta_pruning(child, depth+1, alpha, beta, True, total_points_to_win)
            bestValue = min(bestValue, value)
            beta = min(beta, bestValue)
            if beta <= alpha:
                break
        return bestValue

# Function to determine the winner using alpha-beta pruning
def determine_winner(total_points_to_win, max_points_to_achieve):
    if max_points_to_achieve >= total_points_to_win:
        return "Optimus Prime"
    else:
        return "Megatron"

# Function to shuffle the list of points and determine how many times Optimus Prime wins
def shuffle_and_determine_winner(points, total_points_to_win, s):
    count_optimus_prime_wins = 0
    for i in range(s):
        random.shuffle(points) # Shuffle the list of points
        max_points_to_achieve = alpha_beta_pruning(points, 0, -float('inf'), float('inf'), True, total_points_to_win) # Determine the max points to achieve using alpha-beta pruning
        if determine_winner(total_points_to_win, max_points_to_achieve) == "Optimus Prime":
            count_optimus_prime_wins += 1
    return count_optimus_prime_wins

# Get the student ID from the user
student_id = input("Enter your student ID: ")
if '0' in student_id:
    student_id = student_id.replace('0', '8')
digit_5 = int(student_id[4])
digit_4_3 = int(student_id[2:4])
digit_2_1 = int(student_id[0:2])
min_points = digit_5
max_points = round(digit_4_3 * 1.5)
total_points_to_win = int(str(digit_2_1)[::-1])

# Generate a list of 8 random points between min_points and max_points
points = [random.randint(min_points, max_points) for i in range(8)]

# Task 1: Determine the achieved point by applying alpha-beta pruning and the winner
max_points_to_achieve = alpha_beta_pruning(points, 0, -float('inf'), float('inf'), True, total_points_to_win)
winner = determine_winner(total_points_to_win, max_points_to_achieve)
print("Generated 8 random points between the minimum and maximum point limits:", points)
print("Total points to win:", total_points_to_win)
print("Achieved point by applying alpha-beta pruning =", max_points_to_achieve)
print("The winner is", winner)

# Task 2: Shuffle the list of points s times and determine how many times Optimus Prime wins
s = digit_4_3 % 10
count_optimus_prime_wins = shuffle_and_determine_winner(points, total_points_to_win, s)
```

```python
import random

# function to perform alpha-beta pruning
def alpha_beta(node, alpha, beta, maxPlayer):
    # terminal node, return the points
    if node["level"] == 3:
        return node["points"]
    # max player's turn
    if node["level"] % 2 == 0:
        v = float("-inf")
        for child in node["children"]:
            v = max(v, alpha_beta(child, alpha, beta, False))
            alpha = max(alpha, v)
            if beta <= alpha:
                break
        return v
    # min player's turn
    else:
        v = float("inf")
        for child in node["children"]:
            v = min(v, alpha_beta(child, alpha, beta, True))
            beta = min(beta, v)
            if beta <= alpha:
                break
        return v

# function to generate the game tree
def generate_tree(node, points, maxPlayer):
    # terminal node, set the points
    if node["level"] == 3:
        node["points"] = points.pop(0)
        return
    # max player's turn
    if node["level"] % 2 == 0:
        for i in range(2):
            child = {"level": node["level"] + 1, "children": []}
            node["children"].append(child)
            generate_tree(child, points, False)
    # min player's turn
    else:
        for i in range(4):
            child = {"level": node["level"] + 1, "children": []}
            node["children"].append(child)
            generate_tree(child, points, True)

# function to shuffle the list S times and count Optimus Prime's wins
def shuffle_and_count_wins(points, S, maxPoints, totalPointsToWin):
    wins = 0
    maxWins = 0
    allPoints = []
    for i in range(S):
        random.shuffle(points)
        achievedPoints = alpha_beta({"level": 0, "children": []}, float("-inf"), float("inf"), True)
        allPoints.append(achievedPoints)
        if achievedPoints >= totalPointsToWin:
            wins += 1
        if achievedPoints == maxPoints:
            maxWins += 1
    print("After the shuffle:")
    print("List of all points values from each shuffle:", allPoints)
    print("The maximum value of all shuffles:", maxPoints)
    print("Won", wins, "times out of", S, "number of shuffles")

# main function
def main():
    # input student ID
    studentID = input("Enter your student ID\n")
    # calculate minimum and maximum points
    minPoints = int(studentID[4]) if studentID[4] != "0" else 8
    maxPoints = int(int(studentID[3:5][::-1]) * 1.5)
    totalPointsToWin = int(studentID[-2:])
    # generate random points for the game
    points = [random.randint(10, 120) for i in range(8)]
    print("Generated 8 random points between the minimum and maximum point limits:", points)
    print("Total points to win:", totalPointsToWin)
    # generate game tree and find
main()
```
