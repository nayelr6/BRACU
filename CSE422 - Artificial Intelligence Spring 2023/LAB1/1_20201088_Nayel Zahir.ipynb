```python
file = open('Input file.txt', mode = 'r')

graph_Dict = {}

while True:
  line = file.readline()
  if line == "":
    break
  line = line.split()
  u = line[0]
  hu = int(line[1])
  for i in range(2,len(line)):
    if i % 2 ==0:
      v = line[i]
    else:
      w = line[i]
      if u not in graph_Dict:
        graph_Dict[u] =[]

      graph_Dict[u].append((v,w))




for key in graph_Dict:
  for i in range(len(graph_Dict[key])):
    graph_Dict[key][i] = (graph_Dict[key][i][0], int(graph_Dict[key][i][1]))

#print(graph_Dict)


### File I/O done!!!! 


####search algo 

### Use deque 
# ^^^^^^^^^^^^^^^^^^
### alows me to append into a list from both sides 

#from collections import deque


#no need for deque

class aStar:
    

    def __init__(self, G):
        self.adj_List = G

    def nextNode(self, v):
        return self.adj_List[v]

    # Defining the heuristic values from the values given in the question.
    def heuristic_cost(self, i):
        heuristicDict = {
            'Arad': 366,
            'Bucharest': 0,
            'Craiova': 160,
            'Dobreta': 242,
            'Eforie': 161,
            'Fagaras': 178,
            'Giurgiu': 77,
            'Hirsova': 151,
            'Iasi': 226,
            'Lugoj': 244,
            'Mehadia': 241,
            'Neamt': 234,
            'Oradea': 380,
            'Pitesti': 98,
            'RimnicuVilcea': 193,
            'Sibiu': 253,
            'Timisoara': 329,
            'Urziceni': 80,
            'Vaslui': 199,
            'Zerind': 374
        }

        return heuristicDict[i]

    def aStarSearch(self, s, f):
        visited = set([s])

        traversed = set([])
        totalCost = {s: 0}

        goalCost = {}

        goalCost[s] = 0

        parent_node = {}
        parent_node[s] = s

        while len(visited) > 0:
            Node = None

            for v in visited:
                if Node == None or goalCost[v] + self.heuristic_cost(v) < goalCost[Node] + self.heuristic_cost(Node):
                    Node = v

            if Node == None:
                print('Path does not exist!')
                
            if Node == f:
                pathToTake = []
                cost = 0

                while parent_node[Node] != Node:
                    pathToTake.append(Node)
                    cost += goalCost[Node] - goalCost[parent_node[Node]]
                    Node = parent_node[Node]

                pathToTake.append(s)

                pathToTake.reverse()

                arrow = ' -> '.join(pathToTake)

                print('Path found: {}'.format(arrow))
                
                #print('Path found: {}'.format(pathToTake))
                print('Total cost: {} km'.format(cost))
                return None

            for (node2, weight) in self.nextNode(Node):

                if node2 not in visited and node2 not in traversed:
                    visited.add(node2)
                    parent_node[node2] = Node
                    goalCost[node2] = goalCost[Node] + weight
                else:
                    if goalCost[node2] > (goalCost[Node] + weight):
                        goalCost[node2] = goalCost[Node] + weight
                        parent_node[node2] = Node

                        if node2 in traversed:
                            traversed.remove(node2)
                            visited.add(node2)
            visited.remove(Node)
            traversed.add(Node)

        print('Path does not exist!')
        return None

pathcalc = aStar(graph_Dict)

pathcalc.aStarSearch('Arad', 'Bucharest')


```
